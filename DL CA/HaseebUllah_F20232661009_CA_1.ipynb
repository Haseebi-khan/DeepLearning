the model

# **Exercise 1.06: Using an Optimizer for Simple Linear Regression**

  To introduce the concept of optimization in training models by:

 - Defining a linear regression model and a loss function
 - Using the Adam optimizer to minimize loss and update parameters
 - Observing how model parameters evolve during training to fit the data

image.png

```python
help(tf.optimizers.Adam)
```

```python
w=tf.Variable(0.0)
b=tf.Variable(0.0)
```

```python
x = [1,2,3,4]
y = [0,-1,-2,-3]

def regression(x):
 model=w*x+b
 return model
```

```python
loss = lambda:abs(model.predict(x) - y)
```

```python
optimizer=tf.optimizers.Adam(learning_rate=.01)
```

image.png

```python
for i in range(1000):
  with tf.GradientTape(persistent=True) as tape:

    y_pred = w * x + b

    loss = tf.reduce_mean(tf.square(y - y_pred))

    grads = tape.gradient(loss, [w, b])

    optimizer.apply_gradients(zip(grads, [w, b]))

    if i % 100 == 0:
      print(f"Step {i}: w = {w.numpy():.3f}, b = {b.numpy():.3f}, loss = {loss.numpy():.5f}")
```

# **Activity 1.01: Solving a Quadratic Equation Using an Optimizer**


image.png

```python
import tensorflow as tf
```

```python
x = tf.Variable(0.0)
```

```python
loss = lambda: x**2 - 10*x + 25
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
for step in range(1000):
    with tf.GradientTape() as tape:
        loss = x**2 - 10*x + 25

    grads = tape.gradient(loss, [x])

    optimizer.apply_gradients(zip(grads, [x]))

    if step % 100 == 0:
        print(f"Step {step}: x = {x.numpy():.6f}, loss = {loss.numpy():.6f}")

print("\n Optimized value of x:", x.numpy())
```
